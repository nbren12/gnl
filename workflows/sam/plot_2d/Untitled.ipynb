```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import xarray as xr
```

```python
cld = xr.open_dataarray("cld.nc").isel(time=0)

fig, ax  = plt.subplots(figsize=(6,2))


im= ax.pcolormesh(cld.x/1e6, cld.y/1e6, cld, cmap='Blues_r')
ax.set_aspect(1.0)
plt.colorbar(im)

plt.xlabel('x [1000 km]')
plt.ylabel('y [1000 km]')
plt.title('Cloud Fraction')
plt.savefig("cld.png")
```
