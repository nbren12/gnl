First, we need to use the discrete dynamical model:
$$u_{m+1} = F u_m + \sigma_{m+1}$$

And the observation model:
$$v_{m+1} = g u_{m+1} + \sigma_{m+1}^o$$
The covariance of the observation error is $r = Var(v_{m+1} - g u_{m+1})$ and the model noise has constant variance sigma

Define some functions for the stochastic process and the observation model

```python
def crandn(*args, **kwargs):
    """Complex random normal"""
    return ( 1.0j * randn(*args, **kwargs) + 
     randn(*args, **kwargs) ) /sqrt(2)

def discreteprocess(u0, sig, F):
    """prefect model"""
    u = u0
    while True:
        u = F*u + sig / sqrt(2) * crandn(1)
        yield u
        
def obs(u, g=1.0, r=1.0, **kw):
    """observation model"""
    return g * u + r * crandn(1)

def plotuu(uu):
    fig, (a,b ) = subplots(2)
    a.plot(real(uu))
    a.set_ylabel('Re')
    b.plot(imag(uu))
    b.set_ylabel('Im')
    
from itertools import islice

def getu(nt = 500,  F=1.0, sig=1.0, u0 = 0.0, **kw):
    return array([ u for  u  in 
                  islice( discreteprocess(u0, sig, F), 0, nt)]).ravel()
```

## Kalman filter

The key to getting the kalman filter is to minimize the posterior $ p(u_{m+1} | v_{m+1} ) = p(u_{m+1}) p(v_{m+1} | u_{m+1}) $ , which is the same as maximizing the negative log posterior 

$\newcommand{\ub}{\bar{u}}$
$$ J(u) =  \frac{1}{2} \left(\frac{(u - F\ub_{m+1|m})^2}{r_{m+1|m}} + \frac{(v_{m+1}-gu)^2}{r^o} \right)$$
Here, the prior mean and covariance are known from the forecast models. The first term is a contribution from the model noise  the second term is a contribution from the observation error. The Kalman filter estimates the posterior mean using

$$ \ub_{m+1|m+1} = \text{argmax}_{u} J(u).$$

This is, found using simple calculus:

$$ g\frac{gu - v_{m+1}}{r^o} + \frac{u - \ub_{m+1|m}}{r_{m+1|m}} = 0 $$
$$ u g ( \frac{g}{r^o} +\frac{1}{r_{m+1|m}}) = \frac{v_{m+1}}{r^o} + \frac{ \ub_{m+1|m}}{r_{m+1|m}} = 0 $$

The famous Kalman gain here is the simply the coefficient of the $ v_{m+1} $ in the expression for the posterior mean.

$$ K_{m+1} = \frac{g r_{m+1|m}}{g^2 r_{m+1|m} + r^o}$$

and to summarize we have $\ub_{m+1} =\ub_{m+1|m} + K_{m+1}  ( v_{m+1} - g\ub_{m+1|m})$. After employing the assumption that the forecast noise and observation error are uncorrelated, we can get the update rule for the covariances $r_{m+1|m+1} = (1 - K_{m+1} g ) r_{m+1|m}$.

```python

def kalmanfiltergen(observations,r = 1.0, g= 1.0, F = 1.0, sig=1.0):
    """A generator for the Kalman filtered estimates"""
    ub = 0.0
    rb = 1.0

    for ob in observations:

        # Forecast step
        ub = F *ub
        rb = F * rb * conj(F) + sig

        # Analysis Step
        K = g * rb / ( g**2 * rb + r) # Kalman gain
        ub = ub + K * ( ob - g * ub)
        rb = ( 1- K * g) * rb

        yield ub, rb, K

def kalmanfilter(observations, **kw):
    """Convience function for returning kalman filtered obs"""
    ub, rb,k = zip(*list(kalmanfiltergen(observations, **kw)))
    
    return np.array(ub), np.array(rb), np.array(k)
```

## Results for the perfect model

```python
# Run model and generate obs
d = dict(r = 1.5,
         sig = 1.0,
         F = .2 + .5j,
         g = 1.0)

u = getu(nt=30, **d)
o = [obs(uu, **d) for uu in u]


ub, rb, kal = kalmanfilter(o, **d)

figure(figsize=(8,6))
subplot(221)


plot(real(u), 'k--', label='Truth')
plot(real(o),  'ko', label='Obs')
plot(real(ub), 'k-', label='Filter')
title(r'$\Re{u}$')

subplot(222)
plot(imag(u), 'k--', label='Truth')
plot(imag(o),  'ko', label='Obs')
plot(imag(ub), 'k-', label='Filter')
title(r'$\Im{u}$')
ax = plt.gca()

han, lab= ax.get_legend_handles_labels()
figlegend(han, lab, "upper center", ncol=3, frameon=False)

subplot(223)
plot(kal, label='Kalman gain')
legend(frameon=False)

rms = lambda  x,y : real(((x-y)*conj(x-y)).mean()**0.5)

print "RMS Obs = %s"%(rms(o,u))
print "RMS Filter = %s"%(rms(ub,u))
```
